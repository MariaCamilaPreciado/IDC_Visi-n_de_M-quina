```python
ListImg=[]
Dir= "/content/drive/MyDrive/Visión de Máquina /Entrega/BancoImagenes/Final/0/"
for nameImg in FileList:
    Img = cv2.imread(Dir+nameImg,0)
    ListImg.append(Img)
```

```python
plt.imshow(ListImg[0],cmap='gray')
plt.show()
```

```python
nas = []
for i in range(len(ListImg)):
  ret,thresh4 = cv2.threshold(ListImg[i],127,255,cv2.THRESH_TOZERO)
  nas.append(thresh4)
```

```python
titles = ['Tozero']
newTozero = []
for c in range(len(nas)):
  images = [nas[c]]
  miArray = np.arange(1)
  for i in miArray:
      plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
      plt.title(titles[i])
      plt.xticks([]),plt.yticks([])
      plt.show()
      newTozero.append(images[i])
```

```python
DirNewTozero = '/content/drive/MyDrive/Visión de Máquina /Entrega/BancoImagenes/tozero/0/'
```

```python
for a in range(len(newTozero)):
   cv2.imwrite(DirNewTozero+ str(a)+'_class0' + '.png', newTozero[a])
```
